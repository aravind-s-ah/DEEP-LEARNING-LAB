```
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets import imdb

# Load the IMDB dataset and preprocess it
(x_train, y_train), (x_test, y_test) = keras.datasets.imdb.load_data(num_words=10000)

x_train = pad_sequences(x_train, maxlen=200)
x_test = pad_sequences(x_test, maxlen=200)

# Define a simple sentiment analysis model
model = keras.Sequential([
    keras.layers.Embedding(10000, 128),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

# Compile and train the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print("Training the sentiment model...")
model.fit(x_train, y_train, epochs=3, batch_size=64, verbose=0)
print("Model training complete.")

# Evaluate the model
loss, acc = model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {acc:.4f}")

# Get word index for encoding text
word_index = imdb.get_word_index()

def encode_text(text):
    tokens = text.lower().split()
    # Map words to their integer index, default to 2 for unknown words
    encoded = [word_index.get(word, 2) for word in tokens]
    return pad_sequences([encoded], maxlen=200)

def predict_sentiment(text):
    if not text or text.strip() == "":
        return "Please enter text", 0.0

    data = encode_text(text)
    pred = model.predict(data, verbose=0)[0][0]
    confidence = round(pred * 100, 2)

    if pred > 0.7:
        result = "Positive 😊"
    elif pred < 0.3:
        result = "Negative 😞"
    else:
        result = "Neutral 😐"

    return result, confidence

# Example usage (without Tkinter GUI)
print("\n--- Example Predictions ---")
text1 = "This movie was fantastic, I really enjoyed it!"
sentiment1, conf1 = predict_sentiment(text1)
print(f"Text: '{text1}'")
print(f"Sentiment: {sentiment1} ({conf1}% confidence)")

text2 = "I absolutely hated this film, it was so boring."
sentiment2, conf2 = predict_sentiment(text2)
print(f"Text: '{text2}'")
print(f"Sentiment: {sentiment2} ({conf2}% confidence)")

text3 = "The plot was okay, but the acting could have been better."
sentiment3, conf3 = predict_sentiment(text3)
print(f"Text: '{text3}'")
print(f"Sentiment: {sentiment3} ({conf3}% confidence)")
```
